## Задание 1.4.
### Сформировать матрицу, у которой выше диагонали единицы, а ниже – нули, записать ее в файл, затем считать файл и вывести на экран. Строим массив 28*28:

```python
n = 28
a = np.ones([28, 28])
for i in range(n):
    a[i][i] = 1
    
for i in range(n):
    for j in range(0, i):
        a[i][j] = 0

cv2.imwrite('ris.png', a)
img = cv2.imread('ris.png', 0)
print(img)
cv2.waitKey(0)
plt.imshow(img)
```

## Задание 1.7.
### Считать файл полноцветного изображения cat.jpg, создать для него два места в окне в ширину и два места в высоту. Преобразовать матрицу цветного изображения в полутоновое, из него, используя функцию cv2.threshold, получить бинарное монохромное изображение. Из бинарного монохромного изображения получить его негатив.

```python
img = cv2.imread("image.jpg")
imag = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.subplot(221)
plt.imshow(imag)
plt.axis('off')

gray_img = cv2.imread("image.jpg", 0)
im_bw = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)[1]

plt.subplot(222)
plt.imshow(im_bw,'gray')
plt.axis("off")

im_bwa = cv2.threshold(imag, 128, 255, cv2.THRESH_BINARY)[1]

plt.subplot(223)
plt.imshow(im_bwa)
plt.axis("off")

im_bwb = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY_INV)[1]
plt.subplot(224)
plt.imshow(im_bwb, 'gray')
plt.axis("off")
plt.show()
cv2.waitKey(0)
```

## Задание 1.9. 
### Уменьшить заданное изображение и вывести на печать матрицу уменьшенного изображения. Нам надо сохранить соотношение сторон, чтобы изображение не исказилось при уменьшении. Для этого необходимо вычислить коэффициент уменьшения стороны.


```python
import cv2

img=cv2.imread('car.jpg')
final_wide = 200
r=float(final_wide) / img.shape[1]
dim=(final_wide, int(img.shape[0]*r))
resized=cv2.resize(img,dim,interpolation=cv2.INTER_AREA)

plt.subplot(121)
plt.imshow(resized)
print(resized.shape)
print(resized)

img=cv2.imread('car.jpg', 0)
plt.subplot(122)
plt.imshow(img)
print(img)
```

## Задание 1.10. 
### Считать цветное изображение, конвертировать его в полутоновое, затем получить негатив полутонового изображения.


```python
import cv2

img = cv2.imread("car.jpg", 1)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img = cv2.bitwise_not(img)
plt.imshow(img)
```
